```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import os

current_directory = os.getcwd()
downloads_path = os.path.join(current_directory, 'Downloads', 'AE102_Project_data_file.xlsx')

df = pd.read_excel(downloads_path, usecols=['Transistor_Count_In_Billions', 'Max_Frequency_In_GHz'])


X = df['Transistor_Count_In_Billions'].values.reshape(-1, 1)  # Independent variable
y = df['Max_Frequency_In_GHz'].values                  # Dependent variable

model = LinearRegression()

model.fit(X, y)

coefficient = model.coef_[0]
intercept = model.intercept_

y_pred = model.predict(X)

plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')

plt.xlabel('Transistor Counts')
plt.ylabel('Clock Frequency')
plt.title('Linear Regression')
plt.legend()

plt.show()
coefficient = model.coef_[0]
intercept = model.intercept_

p_value = model.score(X, y)  # Using R-squared as a proxy for p-value

# Print the coefficient and p-value
print('Coefficient:', coefficient)
print('p-value:', p_value)
```
